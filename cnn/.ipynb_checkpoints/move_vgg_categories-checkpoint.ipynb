```python
import pandas as pd
import os
import shutil
import glob
import numpy as np
```

```python
data = pd.read_csv("../Data/final_dank.csv")
```

```python
data.head()
```

```python
imgs = glob.glob("./new raw_data/*")
```

```python
len(imgs)
```

```python
for img in imgs:
    imgID = img.split("\\")[-1].split(".")[0]
    
    try:
        category = np.array(data[data["id"] == imgID]["vgg_feature_1"])[0]
    except:
        continue
    
    category_path = "./vgg_categories/{}".format(category)
    os.mkdir(category_path) if not os.path.isdir(category_path) else None
    shutil.copy(img, category_path)
```

```python
directories = [x[0] for x in os.walk("./vgg_categories/")]
```

```python
len(directories)
```

```python
directories[:5]
```

```python
def getFileNum(directory):
    return len([name for name in os.listdir(directory)])
```

```python
cate_count = {directory.split("/")[-1]:getFileNum(directory) for directory in directories[1:]}
```

```python
cate_count
```

```python
sorted_cate_count = {k: v for k, v in sorted(cate_count.items(), key=lambda item: item[1], reverse = True)}
```
