```python
import numpy as np
from PIL import Image
from io import BytesIO
import glob
import os
```

```python
import tensorflow as tf
import keras
```

```python
# imports for loading and viewing image 
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
```

```python
imageDir = './raw_data/**/*.jpg'
image_files = glob.glob(imageDir, recursive = True)
len(image_files)
```

```python
brokenImgs = []
IMG_DIM = (224, 224)
for file in image_files:
    try:
        img = img_to_array(load_img(file, target_size=IMG_DIM))
    except:
        imgID = file.split('\\')[-1].split('.')[0]
        print("Deleting {}".format(imgID))
        brokenImgs.append(imgID)
        os.remove(file)
```

```python
len(brokenImgs)
```

```python
with open("./brokenImgs", "w") as outfile:
    outfile.write('\n'.join(brokenImgs))
```
