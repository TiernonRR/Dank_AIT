```python
import glob
import shutil
import json
```

```python
files = glob.glob("./data/test/*")
files[0:5]
```

```python
with open("./id_labels.json", "r") as infile:
    dct = json.load(infile)
```

```python
dct
```

```python
for file in files:
    fileName = file.split('\\')[-1]
    fileID = str(fileName.split(".")[0])
    true_label = dct[fileID]['true_label']
    predicted_label = dct[fileID]['predicted_label']
    if (predicted_label == 1):
        if (true_label == 1):
            folder = "True Positives"
        else:
            folder = "False Positives"
    else:
        if (true_label == 1):
            folder = "False Negatives"
        else:
            folder = "True Negatives"
    destPath = "./labels/" + folder + "/" + fileName
    shutil.copy(file, destPath)
```
