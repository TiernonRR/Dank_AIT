```python
import pandas as pd
df_plt = pd.read_json(r'./dank_data_final.json')
```

```python
import pandas as pd
import numpy as np
import json
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn

plt.scatter(df_plt.TextLength, df_plt.ups_normed)
plt.xlabel("Text Length", fontsize=14)
plt.ylabel("Ups(normed)", fontsize=14)
plt.xlim(0,500)
plt.ylim(0,0.04)
plt.title("Relationship between Text Length and Ups(normed)", fontsize=16);
```

```python
plt.scatter(df_plt.Sentiment, df_plt.ups_normed)
plt.xlabel("Sentiment", fontsize=14)
plt.ylabel("Ups(normed)", fontsize=14)
plt.xlim(0,1)
plt.ylim(0,0.04)
plt.title("Relationship between Sentiment and Ups(normed)", fontsize=16);
```

```python
plt.scatter(df_plt.word_count, df_plt.ups_normed)
plt.xlabel("Word count", fontsize=14)
plt.ylabel("Ups(normed)", fontsize=14)
plt.xlim(0,50)
plt.ylim(0,0.04)
plt.title("Relationship between Word_count and Ups(normed)", fontsize=16);
```

```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
cl_words = []
for ind in df_plt.index:     
    cl_words = cl_words + df_plt['processed_words'][ind]
cl_words_str = ' '.join(cl_words) 
wordcloud = WordCloud(
    width = 3500,
    height = 3500,
    background_color = 'black',
    stopwords = STOPWORDS).generate(cl_words_str)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
